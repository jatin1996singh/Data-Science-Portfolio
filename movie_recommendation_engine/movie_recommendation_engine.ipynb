```python
# Import Statements

import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix
from pandas.api.types import CategoricalDtype
```

```python
movies_data = pd.read_csv("movies.csv")
movies_data
```

```python
ratings_data = pd.read_csv("ratings.csv")
ratings_data
```

```python
tags_data = pd.read_csv("tags.csv")
tags_data
```

```python
# user_movies_data = ratings_data.pivot(index = 'movieId', columns = 'userId', values = 'rating').fillna(0)
# user_movies_data


user_c = CategoricalDtype(sorted(ratings_data.userId.unique()), ordered=True)
movie_c = CategoricalDtype(sorted(ratings_data.movieId.unique()), ordered=True)

row = ratings_data.movieId.astype(movie_c).cat.codes
col = ratings_data.userId.astype(user_c).cat.codes

sparse_matrix = csr_matrix((ratings_data["rating"], (row, col)), \
                           shape=(movie_c.categories.size, user_c.categories.size))


print(sparse_matrix)

user_movies_data = pd.DataFrame(
    [ pd.Series(sparse_matrix[i].toarray().ravel()) \
                              for i in np.arange(sparse_matrix.shape[0]) ], \
                         index=movie_c.categories, \
                         columns=user_c.categories)

user_movies_data



```
